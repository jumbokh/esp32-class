```micropython
from machine import Pin
from machine import ADC
import time
p2 = Pin(13, Pin.OUT) # LED: D5 (GPIO14)
adc = ADC(0)          # create ADC object on ADC pin
while True:
     if adc.read() > 500:
            p2.value(1)
     time.sleep(2)
     p2.value(0)
```
